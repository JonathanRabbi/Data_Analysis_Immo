```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/jonathanrabbi/Desktop/Data_Analysis_Immo/CSV_Files/Cat_House_details.csv')

df_House=df[df['Type of property']=='house']

X= df_House[['Bedrooms','Bathrooms','Toilets','Surface of the plot']]

y= df_House['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor

reg=DecisionTreeRegressor(criterion='squared_error',
                          max_depth=10,
                          min_samples_split=10,
                          )
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,y_pred)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(reg,X_train, y_train, cv=10)
```

```python
print(reg.score(X_train,y_train))
print(reg.score(X_test,y_test))
```
